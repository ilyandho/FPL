- Currently I have data for the past 8 seasons. However, I will be using the 22/23 season for testing and training will be on the first 10 GWs of 23/24 season.

- The 23/24 league had 778 players, for whom we will create a row for each (player, gameweek) pair, and add information for the previous four gameweeks as inputs.
The input variables will be as follows:

'name', 'team', 'position', 'gameweek', 'total_points', 'opponent_team', 'transfers_in', 'transfers_out', 'value', 'was_home'

1_total_points,1_assists,1_bonus,1_bps,1_clean_sheets,1_creativity,1_goals_conceded,1_goals_scored,1_ict_index,1_influence

1_yellow_cards,1_goals_scored_team,1_goals_conceded_team,1_was_home,1_transfers_in,1_transfers_out,1_value

1_opp_xG,1_opp_xGA,1_opp_npxG,1_opp_npxGA,1_opp_deep,1_opp_deep_allowed,1_opp_scored,1_opp_missed,1_opp_xpts,1_opp_wins,1_opp_draws,1_opp_loses,1_opp_npxGD,2_opp_xG,2_opp_xGA

Information can be roughly divided in 4 main categories:
- Information before kick-off (which team the player is facing, how many fantasy players transferred him in/out, ...)
- Player-specific data from previous 4 gameweeks (big chances created, points, influence, assists, ...)
- Team specific data from previous 4 gameweeks


The ML model to be used Random Forest with 10_000 decision tress.
